```python
class AD:
    def __init__(self, value, der=1.0):
        self.value = value
        self.der = der

    def __add__(self, other):
        if isinstance(other, AD):
            return AD(self.value + other.value, self.der + other.der)
        else:
            return AD(self.value + other, self.der)

    def __radd__(self, other):
        return self.__add__(other)

    def __sub__(self, other):
        if isinstance(other, AD):
            return AD(self.value - other.value, self.der - other.der)
        else:
            return AD(self.value - other, self.der)

    def __rsub__(self, other):
        if isinstance(other, AD):
            return AD(other.value - self.value, other.der - self.der)
        else:
            return AD(other - self.value, -self.der)

    def __mul__(self, other):
        if isinstance(other, AD):
            return AD(self.value * other.value,
                      self.der * other.value + self.value * other.der)
        else:
            return AD(self.value * other, self.der * other)

    def __rmul__(self, other):
        return self.__mul__(other)
    

    def __repr__(self):
        return f"AD({self.value}, {self.der})"
        
```

```python
import numpy as np
```

```python
def sin(a):
    if isinstance(a, AD):
        return AD(np.sin(a.value), np.cos(a.value) * a.der)
    else:
        return np.sin(a)
```

```python
x = AD(2.0)          # variable 
c = AD(3.0, 0.0)     # constant
```
